```python
import netCDF4 as nc
import numpy as np
```

```python
ds1 = nc.Dataset("ocean_grid_01.nc", mode="a")
```

```python
print(ds1)
```

```python
ds1.grid_type = "regular"
```

```python
geolon_t = ds1["geolon_t"]
geolat_t = ds1["geolat_t"]
```

```python
geolon_t = np.array(geolon_t[:], dtype="float64")
geolat_t = np.array(geolat_t[:], dtype="float64")
```

```python
lon = geolon_t[0, :]
lat = geolat_t[:, 0]
```

```python
grid_x_T = ds1.createVariable("grid_x_T", 'float64', dimensions=("grid_x_T"))
grid_y_T = ds1.createVariable("grid_y_T", 'float64', dimensions=("grid_y_T"))
```

```python
grid_x_T.units = "Degrees_E"
grid_x_T.long_name = "grid_x"
grid_y_T.units = "Degrees_N"
grid_y_T.unites = "grid_y"
```

```python
grid_x_T[:] = lon
grid_y_T[:] = lat
```

```python
ds1.close()
```
